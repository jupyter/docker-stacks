### Embeded R Graphics Commands
- Relevant to be able to quickly examine data. 
- View pairwise graphs to see correlations.
- Histograms to check distributional assumtions. 



```r
#Let's first load the IRIS dataset.
iris<-read.csv(file="../../data/iris.csv", header=TRUE,sep=",")
```

```r
#Now let's check the histogram. 
hist(iris$petalWidth)
hist(iris$sepalLength)
hist(iris$sepalWidth)
```

### Petal Width Seems to Have a Strange Distribution.  Let's separate the species and check again.

```r
iris.setosa<-iris[iris$species=="setosa",]
iris.versicolor<-iris[iris$species=="versicolor",]
iris.virginica<-iris[iris$species=="virginica",]
hist(iris.versicolor$petalWidth)
hist(iris.setosa$petalWidth)
hist(iris.virginica$petalWidth)
```

```r
# Here we can see real separate distiction amon species. 
pairs(iris[1:4], main = "Iris Data", pch = 21, bg = c("red", "green3", "blue")[unclass(iris$species)])
```

```r
#Show Details in petalwidth vs sepal length
plot(iris$petalLength, iris$petalWidth, pch=21, bg=c("red","green3","blue")[unclass(iris$species)], main="Iris Data")
```

### Why ggplot2?

- Great for publication quality visualization.
- Very large ecosystem.
- Very flexible.


```r
# GGplot should be already installed. If it isn't you can install it with 
# install.packages("ggplot2", repos="http://cran.us.r-project.org")
library("ggplot2")
```

```r
#Histogram
ggplot(iris, aes(sepalLength)) + geom_histogram ()
```

```r
install.packages("GGally", repos="http://cran.us.r-project.org")
```

```r
#GGally is an extension to ggplot that enables pairwise. 
library(GGally)
ggpairs(iris, mapping = aes(color = species))
```

```r
#Now let's view one graph with ggplot. 
ggplot(data=iris, aes(x=sepalLength, y=sepalWidth, color=species)) + geom_point(size=3)
```
